```python
from pandas_datareader import data as wb
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ticker = 'TSLA'
start_date = '2021-01-05'
end_date = '2021-01-20'

# Usamos el lector de datos de Alpha Vantage en lugar de Yahoo
data = wb.DataReader(ticker, 'av-daily', start=start_date, end=end_date, api_key='tu_api_key')['close']

returns = data.pct_change()
log_returns = np.log(1 + returns)

# Utiliza displot en lugar de distplot
sns.displot(returns.dropna(), kde=False, bins=20)
plt.xlabel("Retornos Diarios")
plt.ylabel("Frecuencia")
plt.show()
```
